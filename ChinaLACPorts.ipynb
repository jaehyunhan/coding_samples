```python
import pandas as pd 

# https://portwatch.imf.org/datasets/86842d9b6e8948f29200710a8df206f5/about
spillover_country = pd.read_csv("./PortWatch_20250512/Spillover_simulator__country-level_impact.csv")
spillover_country["daily_total_value_at_risk"] = spillover_country["daily_export_value_at_risk"] + spillover_country["daily_import_value_at_risk"]
spillover_country = spillover_country[spillover_country["industry"] == "Total"]


spillover_country_all = spillover_country.copy()
spillover_country = spillover_country[spillover_country["to_iso3"].isin(["CHN", "USA"])]
spillover_country_all
```

```python
spillover_country[(spillover_country["from_iso3"] == "BRA") & (spillover_country["to_iso3"] == "USA")][["from_portname", "from_lat", "from_lon"]].drop_duplicates()
```

```python
portwatch_unique_LACports = pd.read_excel("portwatch_unique_LACports.xlsx")

temp_ports = [p["portwatch_name"] for p in portwatch_unique_LACports.to_dict(orient="records") if not pd.isna(p["mylacports_name"])]
spillover_country_ofinterest = pd.DataFrame([p for p in spillover_country.to_dict(orient="records") if p["from_portname"] in temp_ports])
spillover_country_ofinterest
```

```python
pd.pivot_table(spillover_country[(spillover_country["to_country"] == "United States") & 
               (spillover_country["industry"] == "Total")  &
               (spillover_country["from_country"].isin(["Brazil", "Argentina", "The Bahamas", "Venezuela", "Panama", "Peru", "El Salvador"]))
               ],
              index = "from_country",
              values = "daily_total_value_at_risk",
              aggfunc = ["mean", "median", "sum"])
```

```python
trade_by_country = pd.pivot_table(spillover_country[(spillover_country["to_country"] == "United States") & 
               (spillover_country["industry"] == "Total")],
              index = "from_country",
              values = "daily_total_value_at_risk",
              aggfunc = ["mean", "median", "sum"])
trade_by_country.columns = [f"{agg}_{col}" for agg, col in trade_by_country.columns]
trade_by_country = trade_by_country.reset_index()
#trade_by_country.reset_index()
trade_by_country
```

```python
lac_usa = pd.pivot_table(spillover_country[(spillover_country["to_iso3"]=="USA") & (spillover_country["from_country"].isin(lac_countries))],
               values = "daily_total_value_at_risk",
               index = ["from_portname", "from_country", "from_iso3"],
               aggfunc = "sum")
lac_usa["daily_total_value_at_risk_perc"] = lac_usa["daily_total_value_at_risk"]/sum(lac_usa["daily_total_value_at_risk"])*100
lac_usa = lac_usa.reset_index()
lac_usa
```

```python
spillover_port = pd.read_csv("./PortWatch_20250512/Spillover_simulator%3A_port-level_impact.csv")
spillover_port.head()
```

```python
spillover_port2 = spillover_port[(spillover_port["to_country"] == "United States")][["from_portname", "from_country","to_portname", "average_transit_days", "relative_capacity_at_risk"]]
spillover_port2.head()
```

```python
y_df = spillover_country[spillover_country["to_iso3"] == "USA"][["from_portname", "from_country", "daily_total_value_at_risk"]]
y_df
```

```python
#pd.pivot_table(spillover_port2, values = "daily_capacity_at_risk", index = "from_portname", columns = ["from_country",]

# daily_capacity_at_risk is the amount of maritime capacity being at-risk of facing delays or disruptions 
# due to port disruptions that affect outgoing vessel movements.
# the loaded capacity (payload multiplied by the carrying capacity of the vessel) of vessels on a given route. 
# Values are expressed in Metric Tons.



# calculating the % of at-risk capacity the given port is causing out of U.S. inbound routes
country_totals = spillover_port2.groupby("from_country")["relative_capacity_at_risk"].transform("sum")
spillover_port2["relative_capacity_at_risk_pct"] = spillover_port2["relative_capacity_at_risk"] / country_totals * 100
spillover_port2.head()

# check
#spillover_port2[spillover_port2["from_country"] == "Denmark"]["daily_capacity_at_risk_pct"].sum()

```

```python
pd.pivot_table(spillover_port2[spillover_port2["from_country"].isin(["Brazil", "Argentina", "The Bahamas", "Venezuela", "Panama", "Peru", "El Salvador"])], 
               values = "relative_capacity_at_risk",
               index = "from_country",
               aggfunc = ["mean", "median", "sum"])
```

```python
spillover_port3 = spillover_port2.merge(trade_by_country, left_on = "from_country", right_on = "from_country", how = "outer")
spillover_port3
```

```python
final_df = y_df.merge(spillover_port3, left_on = ["from_portname", "from_country"], right_on = ["from_portname", "from_country"], how = "outer")
final_df.head()
```

```python
to_check_assumptions = final_df.dropna(subset=["daily_capacity_at_risk", "sum_daily_total_value_at_risk", "from_country"])

# Group by from_country and calculate correlation per group
correlation_by_country = (
    to_check_assumptions
    .groupby("from_country")
    .apply(lambda g: g["daily_capacity_at_risk"].corr(g["sum_daily_total_value_at_risk"]))
    .sort_values(ascending=False)
)

pd.DataFrame(correlation_by_country)
pd.DataFrame(correlation_by_country).sort_values(0, ascending=True)
```
